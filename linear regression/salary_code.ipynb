```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv('Salary_dataset.csv')
df
```

```python
plt.ylabel('Salary')
plt.xlabel('Years Experience')
plt.scatter(df.YearsExperience,df.Salary,color='blue',marker='+')
plt.title('Salary vs Experience')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['YearsExperience']],df.Salary)
```

```python
reg.predict(np.array([[7]]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(df[['YearsExperience']],df.Salary)
```

```python
plt.ylabel('Salary')
plt.xlabel('Years Experience')
plt.scatter(df.YearsExperience,df.Salary,color='blue',marker='+')
plt.plot(df.YearsExperience, reg.predict(df[['YearsExperience']]), color='red')
plt.title('Salary vs Experience')
plt.show()
```
