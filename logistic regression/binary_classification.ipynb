```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('couple.csv')
df
```

```python
plt.scatter(df['Age'],df['Married'],marker='+',color='red')
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(df[['Age']],df['Married'],train_size=0.7)
```

```python
print(X_train)
print(X_test)
```

```python
print(y_train)
print(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test)
```
