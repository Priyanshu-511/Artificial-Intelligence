```python
import pandas as pd
```

```python
df = pd.read_csv('people_earning.csv')
df
```

```python
inputs = df.drop('Has_Debt',axis='columns')
target = df['Has_Debt']
print(inputs)
print(target)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_age = LabelEncoder()
le_married = LabelEncoder()
le_income = LabelEncoder()
```

```python
inputs['age_n'] = le_age.fit_transform(inputs['Age'])
inputs['married_n'] = le_married.fit_transform(inputs['Married'])
inputs['income_n'] = le_income.fit_transform(inputs['Income'])
```

```python
inputs.head()
```

```python
input_n = inputs.drop(['Age','Married','Income','Defaulted'], axis='columns')
input_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[1,2,3]])
```
